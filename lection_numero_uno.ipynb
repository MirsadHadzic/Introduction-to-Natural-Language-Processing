```python
# 19. Define a function called vocab_size(text) that has a single parameter for the text, 
# and which returns the vocabulary size of the text.
'''meho = 'anes je meho'

def vocab_size(text):
    return len(text)
vocab_size(meho)'''
def vocab_size(text):
    # Split the text into words
    words = text.split()
    
    # Use a set to get unique words (vocabulary)
    unique_words = set(words)
    
    # Return the size of the vocabulary
    return len(unique_words)

# Example usage:
text_example = "This is an example sentence. This sentence is used for testing the vocab_size function."
result = vocab_size(text_example)

print("Vocabulary Size:", result)
```
