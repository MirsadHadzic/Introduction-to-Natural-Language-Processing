```python
#4. Define a conditional frequency distribution over the Names Corpus that allows
#you to see which initial letters are more frequent for males versus females

#nltk.download('names')
print(nltk.corpus.names.fileids())
nltk.download('names')

distribution = nltk.ConditionalFreqDist(
    [(file, word[0])                          #count words/names first character for each file
     for file in ['female.txt', 'male.txt']
     for word in nltk.corpus.names.words(file)
     ]
  )

distribution.tabulate()
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

def analizirajTekst(text):
    words = word_tokenize(text.lower())
    
    vocabulary = set(words)
    richness = len(vocabulary)/len(words)
    lengthV = len(vocabulary)
    
    fdist = nltk.FreqDist(words[:100])
    
    return vocabulary, richness, lengthV, fdist

analizirajTekst("This is an example sentence. Another sentence follows. Example text for nltk analysis.")
```
