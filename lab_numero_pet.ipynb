```python
#Write regular expressions to match the following classes of strings:
#a. A single determiner (assume that a, an, and the are the only determiners)
#b. An arithmetic expression using integers, addition, and multiplication, such as 2*3+8

import re
import nltk
re_a = r'(\ban?\b|\bthe\b)'
re_b = r'[\d\*\+]+'

# Test the regular expressions on sample texts
text_a = 'A single determiner (assume that a, an, and the are the only determiners)'
text_b = 'An arithmetic expression using integers, addition, and multiplication, such as 10*3+8'

# Find matches using re.findall
matches_a = re.findall(re_a, text_a)
matches_b = re.findall(re_b, text_b)

# Print the results
print("Determiners:", matches_a)
print("Arithmetic Expressions:", matches_b)

f = nltk.FreqDist(matches_a)
f
#\b matches any character that starts with 'a' and ends in 'a' or 'an'
#? means match a letter 'n' 0 or 1 time
#| means or
#\b matches any character that is 'the'
#\d matches one digit
#\* matches *
#\+ matches +
#[] means subgroup
#+ means that group/character will repeat one or more times
#https://www.computerhope.com/unix/regex-quickref.html
```

```python
#5. Save some text into a file corpus.txt. Define a function load(f) that reads from the file named in its sole argument,
#and returns a string containing the text of the file.
#a. Use nltk.regexp_tokenize() to create a tokenizer that tokenizes the various kinds of punctuation in this text.
#Use one multiline regular expression inline comments, using the verbose flag (?x).
#b. Use nltk.regexp_tokenize() to create a tokenizer that tokenizes the following kinds of expressions: monetary
#amounts; dates; names of people and organizations.

def save_text_to_file(text, filename = 'C:/Users/Anes/Downloads/randomTxtAssignmentTask.txt'):
    with open(filename, 'w', encoding='utf-8') as file:
        file.write(text)
def load(f):
    with open(f, 'r', encoding= 'utf-8') as file:
        text = file.read()
    return text
# Sample text
sample_text = """
This is a sample text with various punctuation marks. How are you doing today?
Let's tokenize this text and see how it works! Monetary expressions like $100.50, dates like 12/31/2023, and names such as John Doe and ACME Inc. should be recognized.
"""

# Save the sample text to a file
save_text_to_file(sample_text)

# Task 2a: Tokenize various kinds of punctuation
def tokenize_punctuation(text):
    # Multiline regular expression with inline comments
    punctuation_pattern = r'''(?x)
        \w+             # Match words
      | [.,!?;]         # Match common punctuation
      | ['"“”‘’']       # Match different types of quotation marks
      | [\(\)\[\]\{\}]   # Match various types of brackets
    '''
    return nltk.regexp_tokenize(text, punctuation_pattern)

# Tokenize punctuation in the sample text
punctuation_tokens = tokenize_punctuation(sample_text)
print("Tokens with Punctuation:", punctuation_tokens)

# Task 2b: Tokenize monetary amounts, dates, and names
def tokenize_expressions(text):
    # Multiline regular expression with inline comments
    expression_pattern = r'''(?x)
        \$[\d,]+(?:\.\d{2})?       # Monetary amounts like $100.00
      | \d{1,2}/\d{1,2}/\d{2,4}     # Dates like 12/31/2023 or 12/31/23
      | [A-Z][a-z]+                 # Names of people and organizations
    '''
    return nltk.regexp_tokenize(text, expression_pattern)

# Tokenize expressions in the sample text
expression_tokens = tokenize_expressions(sample_text)
print("Monetary Amounts, Dates, and Names:", expression_tokens)

```

For this task find a web page like https://www.bbc.co.uk/news/uk-england-67271027 and load that page as html. You will have all tags and raw code of the page to process.
Sometimes data in tags can contain key words for the NLP. Eq. ALT in IMG tag can contribute to NLP.
That is why we will use regex to extract those words and sentences. To see regex expression you can use https://regex101.com/r/cU5lC2/1.

You can do multiple of regex extraction on same text to get all words. Eq. you can first find words from P, H1 tag then ALT from IMG tag. Since in this website we do not have text in DIV tags we often have another tag in it that we need to also extract wtih regex.
After we are satisfied with extracted text we can process that text, tokenize it, do frequency distribution, see most common stop words, remove stop words, transform it to lowercase words, then also do frequency distribution.

```python
import requests
from urllib import request
import re
from bs4 import BeautifulSoup
from nltk.corpus import stopwords
from nltk import FreqDist
import matplotlib.pyplot as plt

# Function to fetch HTML content from a URL
def get_html_content(url):
    response = requests.get(url)
    return response.text if response.status_code == 200 else None

# Function to extract text from specific HTML tags using regex
def extract_text_with_regex(html_content, tag):
    pattern = re.compile(f'<{tag}.*?>(.*?)</{tag}>', re.DOTALL)
    matches = re.findall(pattern, html_content)
    return matches

# Function to tokenize text and create frequency distribution
def process_text(text):
    tokens = re.findall(r'\b\w+\b', text.lower())  # Tokenize and convert to lowercase
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word not in stop_words]
    fdist = FreqDist(filtered_tokens)
    return fdist

# Example URL
url = 'https://www.bbc.co.uk/news/uk-england-67271027'

# Fetch HTML content
html_content = get_html_content(url)

if html_content:
    # Extract text from specific tags using regex
    p_text = extract_text_with_regex(html_content, 'p')
    h1_text = extract_text_with_regex(html_content, 'h1')
    img_alt_text = extract_text_with_regex(html_content, 'img alt')
    div_text = extract_text_with_regex(html_content, 'div')

    # Combine extracted text
    combined_text = ' '.join(p_text + h1_text + img_alt_text + div_text)

    # Process text, tokenize, and create frequency distribution
    fdist = process_text(combined_text)

    # Display the most common words
    print("Most Common Words:")
    print(fdist.most_common(10))

    # Plot the frequency distribution
    fdist.plot(30, cumulative=False)
    plt.show()
else:
    print("Failed to fetch HTML content.")
```
