```python
#14. Find all the four-letter words in the Chat Corpus (text5).
# With the help of a frequency distribution (FreqDist), show these words in decreasing order of frequency.

# freq_dist = FreqDist(text5)
meho = [w for w in text5 if len(w)==4 and w.isalpha()]
freq_dist = FreqDist(meho)
#print(list(freq_dist))
print(freq_dist.most_common(10))
print(freq_dist.plot(10))
print(freq_dist.most_common()[-10:][::-1])
```
