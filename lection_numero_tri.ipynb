```python
# Two distinct words having the same spelling are called homonyms.
# saw and saw

def unusual_words(text):
    text_vocab = set(w.lower() for w in text if w.isalpha())
    english_vocab = set(w.lower() for w in nltk.corpus.words.words())
    unusual = text_vocab.difference(english_vocab)
    return sorted(unusual)
```

```python
def content_fraction(text):
    stopwords = nltk.corpus.stopwords.words('english')
    content = [w for w in text if w.lower() not in stopwords]
    return len(content)/len(text)

print(content_fraction(nltk.corpus.reuters.words()))
```

```python
# Define a function find_language() that takes a string as its argument and returns 
# a list of languages that have that string as a word. Use the udhr corpus and limit your searches 
# to files in the Latin-1 encoding.

# import nltk
# nltk.download('udhr')
# nltk.corpus.udhr.fileids()
# from nltk.corpus import udhr

def find_language(string):
    languages = []
    target_languages = []

    for name in nltk.corpus.udhr.fileids():
        if 'Latin1' in name:
            languages.append(name)

    for lang in languages:
        if string in udhr.words(lang):
            target_languages.append(lang)

    return target_languages
```
