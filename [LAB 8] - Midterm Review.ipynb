```python
#(48 pts) You will be working on one of the Gutenberg project's text - austen-persuasion.txt. Use NLTK library to finish following tasks:
#Load text

import nltk
nltk.download('gutenberg')

text = nltk.corpus.gutenberg.raw("austen-persuasion.txt")
```

```python
#Find the total number of characters, words, and sentences (using NLTK)

nltk.download('punkt')

words = nltk.corpus.gutenberg.words('austen-persuasion.txt')
sents = nltk.corpus.gutenberg.sents('austen-persuasion.txt')

print("no. of characters: ", len(text))
print("no. of words: ", len(words))
print("no. of words: ", len(sents))
```

```python
#Find the total number of distinct words

print("no. of distinct words: ", len(set(words)))
```

```python
#Find the frequency distribution of the text and the most common words

fdist = nltk.FreqDist(words)
fdist.most_common(10)
```

```python
#Find how many times the word „love“ appears

fdist = nltk.FreqDist(words)
fdist.get('love') # fdist['love']
```

```python
#Generate a cumulative frequency plot for the 8 most common words.

fdist.plot(8)
```

```python
#Tokenize the text and show the first 14 tokens.

tokenized = nltk.word_tokenize(text)
tokenized[:14]
```

```python
#Find part-of-speech tags for those tokens

#nltk.download('averaged_perceptron_tagger')

tags = nltk.pos_tag(tokenized[:14])
tags
```

```python
#Find the 50 most frequent bigrams of a text, omitting bigrams that contain stopwords.

#nltk.download('stopwords')


stop_words = nltk.corpus.stopwords.words('english')

text_no_stop_words = [word for word in words if word.lower() not in stop_words]

bigrams = nltk.bigrams(text_no_stop_words)
fdistBigram = nltk.FreqDist(bigrams)

fdistBigram.most_common(50)
```

```python
#Use the Porter Stemmer to normalize tokenized text, calling the stemmer on each word.

porter = nltk.PorterStemmer()

def word_stemm(text):
  porter_words  = [porter.stem(w) for w in text]
  print(porter_words)

word_stemm(words)
```

```python
#Use WordNet Lemmatizer to find a list of valid lemmas

#nltk.download('wordnet')
lemma = nltk.WordNetLemmatizer()

def word_lemma(text):
  lancaster_words = [lemma.lemmatize(w) for w in text]
  print(lancaster_words)

word_lemma(words)
```

```python
#(25 pts) Define a function find_language() that takes a string as its argument and returns a list of languages that have that string as a word.
#Use the udhr corpus and limit your searches to files in the Latin-1 encoding.

nltk.download('udhr')
from nltk.corpus import udhr

def find_language(string):
    languages = []
    target_languages = []

    for name in nltk.corpus.udhr.fileids():
        if 'Latin1' in name:
            languages.append(name)

    for lang in languages:
        if string in nltk.corpus.udhr.words(lang):
            target_languages.append(lang)

    return target_languages

print(find_language("DEKLARACIJA"))

nltk.corpus.udhr.words('Serbian_Srpski-Latin2')
```

```python
#(27 pts) You will be working with the NLTK Brown Corpus, which contains text samples from various genres.
#Your objective is to analyze the distribution of word lengths in different genres. Implement the following tasks:
#Define a function genre_word_length_distribution(corpus, genres) that takes the Brown Corpus and a list of genres as input.
#The function should return a Conditional Frequency Distribution where the conditions are genres,
#and the values are the lengths of words in the corresponding genre.

nltk.download('brown')

def genre_word_length_distribution(corpus, genres):
  distribution = nltk.ConditionalFreqDist(
    [(genre, len(word))
     for genre in genres
     for word in corpus.words(categories = genre)
     ]
  )
  return distribution
```

```python
#Use the function to generate the Conditional Frequency Distribution for a specific set of genres, e.g., ['news', 'romance', 'science_fiction'].
#Plot the Conditional Frequency Distribution to visualize the distribution of word lengths across the selected genres.

genres = ['news', 'romance', 'science_fiction']

genre_word_length_distribution(nltk.corpus.brown, genres).plot()
```
